### geopy

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import googlemaps
```

```python
def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    geo = geolocoder.geocode(address)
    crd = {"위도": str(geo.latitude), "경도": str(geo.longitude)}

    return crd

crd = geocoding("서울특별시 종로구 창성동 109-6")
print(crd['위도'])
print(crd['경도'])
```

```python
df = pd.read_csv("./seoul_closed_clothes_final.csv", encoding='utf-8')
```

```python
geolocator = Nominatim(user_agent="geoapiExercies")

#주소로 부터 위도와 경도 찾는 함수
def get_la_lo(address):
    try:
        location = geolocator.geocode(address, timeout=10)
        if location:
             return location.latitude, location.longitude
        else:
             return None, None
    except GeocoderTimedOut:
        return get_la_lo(address)   
```

##### 위도, 경도 1차 변환

```python
#위도와 경도가 없는 경우 업데이트

for row in df.itertuples():
    if pd.isna(row.위도) or pd.isna(row.경도):
        latitude, longitude = get_la_lo(row.주소)
        df.at[row.Index, '위도'] = latitude
        df.at[row.Index, '경도'] = longitude
        
print(df)
```
