#### 1. 강남구

```python
df_gangnam = pd.read_csv("./file/서울특별시 강남구_의류수거함 위치 현황_20240311.csv",encoding='cp949')

# 구 칼럼 추가해서 모두 '강남구'로 채우기
df_gangnam['구'] = '강남구'

# 주소를 공백 기준으로 분리했고, 동 이름을 동 칼럼 으로 생성
dong = []
for i in range(len(df_gangnam)):
    dong.append(df_gangnam['지번주소'].str.split(' ')[i][2])
df_gangnam['동'] = dong

#칼럼 삭제
df_gangnam.drop(['연번','지번주소'],axis=1,inplace=True)

# 정규식을 이용해 도로명 주소에서 (00동) 을 지우고 , 주소라는 칼럼 생성
df_gangnam['주소'] = df_gangnam['도로명 주소'].str.replace(r' \([가-힣0-9]+동(?:, [가-힣0-9]+)?\)', '', regex=True)

#칼럼 순서 변경
df_gangnam = df_gangnam[['구','동','주소','위도','경도']]
```

#### 2. 강동구  

```python
df_gangdong = pd.read_excel("./file/서울특별시 강동구 의류수거함 위치정보.xlsx")

# 첫 번째 행을 칼럼 이름으로 설정
df_gangdong.columns = df_gangdong.iloc[0]

# 첫 번째 행을 삭제하고 인덱스를 재설정
df_gangdong = df_gangdong[1:].reset_index(drop=True)

# '동명' 칼럼명 '동' 으로 변경
df_gangdong = df_gangdong.rename(columns={'동명': '동'})


# # 구 칼럼 추가해서 모두 '강동구'로 채우기
df_gangdong['구'] = '강동구'

#주소 합치기
df_gangdong['전체주소'] = df_gangdong['구'] + ' ' + df_gangdong['동'] + ' ' + df_gangdong['주소'].astype(str)

# 연번, 유형 칼럼 삭제
df_gangdong.drop(['연번', '유형','주소'], axis=1, inplace=True)

# 칼럼명 변경
df_gangdong = df_gangdong.rename(columns={'전체주소': '주소'})


#칼럼 순서 변경
df_gangdong = df_gangdong[['구','동','주소']]
```

#### 7. 구로구 


```python
df_guro = pd.read_csv("./file/서울특별시 구로구_의류수거함_20240607.csv", encoding='cp949')

# 결측치 행 제거
df_guro.dropna(axis='index',inplace=True)

# 칼럼 삭제
df_guro.drop(['연번','데이터기준일자'],axis=1, inplace=True)

# 접두사
prefix = '서울특별시 구로구 '
pattern = r'^서울특별시 구로구'

def add_prefix(주소):
    if not 주소.startswith('서울특별시 구로구'):
        return prefix + 주소
    else:
        return 주소
df_guro['주소'] = df_guro['주소'].apply(add_prefix)   

#주소 잘못된 값 수정
df_guro['주소'] = df_guro['주소'].replace('서울특별시 구로구 서울특별기 구로구 경인로35길64-18', '서울특별시 구로구 경인로35길64-18')

 
# 칼럼명 변경
df_guro = df_guro.rename(columns={'행정동': '동'})

# 구 칼럼 추가해서 모두 '구로구'로 채우기
df_guro['구'] = '구로구'

#칼럼 순서 변경
df_guro = df_guro[['구','동','주소']]
```
